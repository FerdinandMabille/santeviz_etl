```python
import pandas as pd
```

```python
file = "snatifsejoursssr.csv"
data = pd.read_csv(file, sep=";", low_memory = False)
df = pd.DataFrame(data)
df
```

```python
df = df.drop_duplicates(subset=['SEJ_NUM'], keep ='first')
df.reset_index(inplace = True, drop = True)
df
```

```python
i = 0
    
for elem in df['EXE_SOI_DTD']:
    if i < 2966 :
        if elem[2] == '/':
            print("old :", df['EXE_SOI_DTD'][i])
            df['EXE_SOI_DTD'][i] =  df['EXE_SOI_DTD'][i][0:10]
            print("new :", df['EXE_SOI_DTD'][i])

        else:
            print("old :", df['EXE_SOI_DTD'][i])
            
            day = df['EXE_SOI_DTD'][i][0:2]
            if elem[2:5] == "JAN":
                month = "01"
            elif elem[2:5] == "FEB":
                month = "02"
            elif elem[2:5] == "MAR":
                month = "03"
            elif elem[2:5] == "APR":
                month = "04"
            elif elem[2:5] == "MAY":
                month = "05"
            elif elem[2:5] == "JUN":
                month = "06"
            elif elem[2:5] == "JUL":
                month = "07"
            elif elem[2:5] == "AUG":
                month = "08"
            elif elem[2:5] == "SEP":
                month = "09"
            elif elem[2:5] == "OCT":
                month = "10"
            elif elem[2:5] == "NOV":
                month = "11"
            elif elem[2:5] == "DEC":
                month = "12"
            year = elem[5:9]

            df['EXE_SOI_DTD'][i] = day + "/" + month + '/' + year
            print("new :", df['EXE_SOI_DTD'][i])
        
        i = i + 1
```
